```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ƒê·ªçc d·ªØ li·ªáu
product = pd.read_csv("product.csv")

# Lo·∫°i b·ªè d√≤ng Total v√† d√≤ng c√≥ Year = NaN
product = product[product['Year'].notna()]
product = product[product['Year'] != 'Total']

# Chuy·ªÉn Year th√†nh int
product['Year'] = product['Year'].astype(int)

# Danh s√°ch c√°c c·ªôt s·ªë c·∫ßn chuy·ªÉn ƒë·ªïi (b·ªè d·∫•u ph·∫©y v√† chuy·ªÉn sang float)
cols_to_convert = [
    'Sum of product_price', 'Sum of product_cost', 'Sum of FixedCost',
    'Sum of flexible_fee', 'Sum of commission',
    'Gross Profit', 'Sum of ProfitOrder'
]

for col in cols_to_convert:
    product[col] = product[col].astype(str).str.replace(',', '').astype(float)

# T·∫°o c·ªôt Gross Profit ngh√¨n t·ª∑
product['Gross Profit (ngh√¨n t·ª∑)'] = product['Gross Profit'] / 1e12

# B·∫£ng s·ªë li·ªáu Gross Profit theo nƒÉm v√† s·∫£n ph·∫©m
gross_profit_table = product.pivot(index='Year', columns='product_profile', values='Gross Profit (ngh√¨n t·ª∑)').round(2)
print("üìä B·∫£ng Gross Profit theo s·∫£n ph·∫©m v√† nƒÉm (ƒê∆°n v·ªã: ngh√¨n t·ª∑ VND):")
display(gross_profit_table)

# Chu·∫©n b·ªã d·ªØ li·ªáu v·∫Ω bi·ªÉu ƒë·ªì
gross_profit_plot = gross_profit_table.reset_index().melt(id_vars='Year', var_name='product_profile', value_name='Gross Profit (ngh√¨n t·ª∑)')

# V·∫Ω bi·ªÉu ƒë·ªì
plt.figure(figsize=(10, 6))
sns.barplot(data=gross_profit_plot, x='Year', y='Gross Profit (ngh√¨n t·ª∑)', hue='product_profile')
plt.title('Gross Profit theo t·ª´ng s·∫£n ph·∫©m (ƒê∆°n v·ªã: ngh√¨n t·ª∑ VND)')
plt.ylabel('Gross Profit (ngh√¨n t·ª∑)')
plt.xlabel('NƒÉm')
plt.legend(title='S·∫£n ph·∫©m')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ƒê·ªçc d·ªØ li·ªáu
order = pd.read_csv("order.csv")

# Chuy·ªÉn order_date sang datetime
order['order_date'] = pd.to_datetime(order['order_date'])

# T·∫°o c·ªôt Year
order['year'] = order['order_date'].dt.year

# L·ªçc c√°c ƒë∆°n h√†ng ƒë√£ giao th√†nh c√¥ng
order_success = order[order['order_status'] == 'successfully delivered']

# ƒê·∫øm s·ªë ƒë∆°n theo nƒÉm v√† product_profile (ch·ªâ 3 lo·∫°i A, B, C)
product_counts = (
    order_success[order_success['product_profile'].isin(['product_A', 'product_B', 'product_C'])]
    .groupby(['year', 'product_profile'])
    .size()
    .reset_index(name='quantity')
)

# Pivot th√†nh b·∫£ng s·ªë li·ªáu
product_table = product_counts.pivot(index='year', columns='product_profile', values='quantity').fillna(0).astype(int)
print("üìä B·∫£ng s·ªë l∆∞·ª£ng s·∫£n ph·∫©m theo nƒÉm:")
display(product_table)

# V·∫Ω bi·ªÉu ƒë·ªì
plt.figure(figsize=(10, 6))
sns.barplot(data=product_counts, x='year', y='quantity', hue='product_profile')
plt.title('S·ªë l∆∞·ª£ng s·∫£n ph·∫©m theo t·ª´ng nƒÉm (Ch·ªâ t√≠nh ƒë∆°n h√†ng th√†nh c√¥ng)')
plt.ylabel('S·ªë l∆∞·ª£ng s·∫£n ph·∫©m (ƒë∆°n h√†ng)')
plt.xlabel('NƒÉm')
plt.legend(title='S·∫£n ph·∫©m')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# ƒê·ªçc d·ªØ li·ªáu
sales_channel = pd.read_csv("sales_channel.csv")
sales_channel = sales_channel[sales_channel['Year'].notna()]
sales_channel = sales_channel[sales_channel['Year'] != 'Total']
sales_channel['Year'] = sales_channel['Year'].astype(int)

# L√†m s·∫°ch v√† chuy·ªÉn ƒë·ªïi s·ªë
cols = ['Sum of FixedCost', 'Sum of flexible_fee', 'Sum of commission',
        'Gross Profit', 'Profit after expenses', 'Net Profit']
for col in cols:
    sales_channel[col] = sales_channel[col].astype(str).str.replace(',', '').astype(float)

# üëâ M√¥ ph·ªèng: gi·∫£m 10% flexible_fee v√† 20% commission
sales_channel['Sim_FlexibleFee'] = sales_channel['Sum of flexible_fee'] * 0.9
sales_channel['Sim_Commission'] = sales_channel['Sum of commission'] * 0.8

# üëâ T√≠nh l·∫°i chi ph√≠ sau m√¥ ph·ªèng
sales_channel['Sim_TotalCost'] = (
    sales_channel['Sum of FixedCost'] +
    sales_channel['Sim_FlexibleFee'] +
    sales_channel['Sim_Commission']
)

# üëâ T√≠nh l·∫°i Profit after expenses m√¥ ph·ªèng:
sales_channel['Sim_ProfitAfter'] = sales_channel['Gross Profit'] - sales_channel['Sim_TotalCost']

# üëâ Net Profit m√¥ ph·ªèng = ProfitAfter + c√°c kho·∫£n l·ª£i nhu·∫≠n v·∫≠n h√†nh kh√°c kh√¥ng thay ƒë·ªïi
# => Ta gi·∫£ ƒë·ªãnh thay ƒë·ªïi ch·ªâ t·ª´ ph√≠ n√†y => delta gi·ªØa ProfitAfter m·ªõi v√† c≈©
sales_channel['Delta_ProfitAfter'] = sales_channel['Sim_ProfitAfter'] - sales_channel['Profit after expenses']
sales_channel['Sim_NetProfit'] = sales_channel['Net Profit'] + sales_channel['Delta_ProfitAfter']

# ƒê∆°n v·ªã ngh√¨n t·ª∑ ƒë·ªÉ d·ªÖ ƒë·ªçc
sales_channel['Net Profit (g·ªëc)'] = sales_channel['Net Profit'] / 1e12
sales_channel['Net Profit (m√¥ ph·ªèng)'] = sales_channel['Sim_NetProfit'] / 1e12
sales_channel['TƒÉng l·ª£i nhu·∫≠n (ngh√¨n t·ª∑)'] = sales_channel['Net Profit (m√¥ ph·ªèng)'] - sales_channel['Net Profit (g·ªëc)']

# K·∫øt qu·∫£ t√≥m t·∫Øt
summary = sales_channel.groupby('Year')[
    ['Net Profit (g·ªëc)', 'Net Profit (m√¥ ph·ªèng)', 'TƒÉng l·ª£i nhu·∫≠n (ngh√¨n t·ª∑)']
].sum().round(2)

print("üìä K·∫øt qu·∫£ m√¥ ph·ªèng khi c·∫Øt gi·∫£m 10% flexible_fee v√† 20% commission:")
display(summary)
```

```python
import pandas as pd

# ƒê·ªçc v√† x·ª≠ l√Ω d·ªØ li·ªáu
sales = pd.read_csv("sales_channel.csv")
sales = sales[sales['Year'].notna()]
sales = sales[sales['Year'] != 'Total']
sales['Year'] = sales['Year'].astype(int)

# Chuy·ªÉn c√°c c·ªôt t√†i ch√≠nh v·ªÅ d·∫°ng float
cols = ['Sum of product_price', 'Sum of product_cost', 'Sum of FixedCost',
        'Sum of flexible_fee', 'Sum of commission',
        'Gross Profit', 'Profit after expenses', 'Net Profit']
for col in cols:
    sales[col] = sales[col].astype(str).str.replace(',', '').astype(float)

# Copy b·∫£ng ƒë·ªÉ m√¥ ph·ªèng
sim_sales = sales.copy()

# M√¥ ph·ªèng tƒÉng 15% cho website
website_mask = sim_sales['sales_channel'] == 'website'
sim_sales.loc[website_mask, 'Sum of product_price'] *= 1.15
sim_sales.loc[website_mask, 'Sum of product_cost'] *= 1.15
sim_sales.loc[website_mask, 'Sum of FixedCost'] *= 1.15
sim_sales.loc[website_mask, 'Sum of flexible_fee'] *= 1.15
sim_sales.loc[website_mask, 'Sum of commission'] *= 1.15
sim_sales.loc[website_mask, 'Gross Profit'] *= 1.15
sim_sales.loc[website_mask, 'Profit after expenses'] *= 1.15
sim_sales.loc[website_mask, 'Net Profit'] *= 1.15

# M√¥ ph·ªèng gi·∫£m 15% cho stores
stores_mask = sim_sales['sales_channel'] == 'stores'
sim_sales.loc[stores_mask, 'Sum of product_price'] *= 0.85
sim_sales.loc[stores_mask, 'Sum of product_cost'] *= 0.85
sim_sales.loc[stores_mask, 'Sum of FixedCost'] *= 0.85
sim_sales.loc[stores_mask, 'Sum of flexible_fee'] *= 0.85
sim_sales.loc[stores_mask, 'Sum of commission'] *= 0.85
sim_sales.loc[stores_mask, 'Gross Profit'] *= 0.85
sim_sales.loc[stores_mask, 'Profit after expenses'] *= 0.85
sim_sales.loc[stores_mask, 'Net Profit'] *= 0.85

# T√≠nh l·∫°i t·ªïng l·ª£i nhu·∫≠n t·ª´ng nƒÉm (tr∆∞·ªõc & sau m√¥ ph·ªèng)
summary = pd.DataFrame()
summary['Net Profit g·ªëc'] = sales.groupby('Year')['Net Profit'].sum() / 1e12
summary['Net Profit m√¥ ph·ªèng'] = sim_sales.groupby('Year')['Net Profit'].sum() / 1e12
summary['Ch√™nh l·ªách (ngh√¨n t·ª∑)'] = summary['Net Profit m√¥ ph·ªèng'] - summary['Net Profit g·ªëc']

print("üìä T√°c ƒë·ªông khi TƒÇNG ƒë·∫ßu t∆∞ v√†o website 15% v√† GI·∫¢M ƒë·∫ßu t∆∞ v√†o stores 15%:")
display(summary.round(2))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ƒê·ªçc d·ªØ li·ªáu
order = pd.read_csv("order.csv")
order["order_date"] = pd.to_datetime(order["order_date"])
order = order[
    (order["order_status"] == "successfully delivered") &
    (order["product_profile"] == "product_C")
]

# T·∫°o c·ªôt qu√Ω
order["quarter"] = order["order_date"].dt.to_period("Q").astype(str)

# L·∫•y kh√°ch h√†ng theo qu√Ω
customer_by_quarter = (
    order.groupby("quarter")["customer_id"]
    .unique()
    .reset_index(name="customers")
)

# T√≠nh Retention Rate
retention_data = []
for i in range(1, len(customer_by_quarter)):
    prev_customers = set(customer_by_quarter.loc[i-1, "customers"])
    curr_customers = set(customer_by_quarter.loc[i, "customers"])

    if len(prev_customers) == 0:
        retention = None
    else:
        retained = len(prev_customers & curr_customers)
        retention = retained / len(prev_customers)

    retention_data.append({
        "quarter": customer_by_quarter.loc[i, "quarter"],
        "retention_rate": retention
    })

retention_df = pd.DataFrame(retention_data)

# ƒê·ªãnh d·∫°ng hi·ªÉn th·ªã %
retention_df["retention_rate_display"] = (retention_df["retention_rate"] * 100).round(2).astype(str) + "%"

# Hi·ªán b·∫£ng s·ªë li·ªáu
print("T·ª∑ l·ªá duy tr√¨ s·ª≠ d·ª•ng s·∫£n ph·∫©m C theo qu√Ω:")
print(retention_df[["quarter", "retention_rate_display"]].to_string(index=False))

# Bi·ªÉu ƒë·ªì
plt.figure(figsize=(10,6))
plt.plot(retention_df["quarter"], retention_df["retention_rate"], marker='o', color='green')
plt.title("T·ª∑ l·ªá duy tr√¨ s·ª≠ d·ª•ng s·∫£n ph·∫©m C theo qu√Ω")
plt.xlabel("Qu√Ω")
plt.ylabel("Retention Rate")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ƒê·ªçc d·ªØ li·ªáu
product = pd.read_csv("product.csv")

# Lo·∫°i b·ªè d√≤ng t·ªïng
product = product[product["Year"].notna()]
product["Year"] = product["Year"].astype(int)

# Lo·∫°i b·ªè ƒë·ªãnh d·∫°ng s·ªë ki·ªÉu chu·ªói c√≥ d·∫•u ph·∫©y
for col in ['Sum of product_price', 'Gross Profit', 'Sum of ProfitOrder']:
    product[col] = product[col].astype(str).str.replace(",", "").astype(float)

# T·ªïng theo nƒÉm (t·∫•t c·∫£ s·∫£n ph·∫©m)
total_by_year = (
    product.groupby("Year")[["Sum of product_price", "Gross Profit", "Sum of ProfitOrder"]]
    .sum()
    .reset_index()
    .rename(columns={
        "Sum of product_price": "total_price",
        "Gross Profit": "total_gross",
        "Sum of ProfitOrder": "total_profit"
    })
)

# D·ªØ li·ªáu ri√™ng s·∫£n ph·∫©m C
product_C = product[product["product_profile"] == "product_C"]
c_by_year = (
    product_C.groupby("Year")[["Sum of product_price", "Gross Profit", "Sum of ProfitOrder"]]
    .sum()
    .reset_index()
    .rename(columns={
        "Sum of product_price": "C_price",
        "Gross Profit": "C_gross",
        "Sum of ProfitOrder": "C_profit"
    })
)

# G·ªôp ƒë·ªÉ t√≠nh t·ª∑ tr·ªçng
df = pd.merge(total_by_year, c_by_year, on="Year")

df["T·ª∑ tr·ªçng doanh thu C (%)"] = df["C_price"] / df["total_price"] * 100
df["T·ª∑ tr·ªçng gross profit C (%)"] = df["C_gross"] / df["total_gross"] * 100
df["T·ª∑ tr·ªçng l·ª£i nhu·∫≠n sau chi ph√≠ C (%)"] = df["C_profit"] / df["total_profit"] * 100

# Hi·ªÉn th·ªã b·∫£ng s·ªë li·ªáu
print("T·ª∑ tr·ªçng ƒë√≥ng g√≥p c·ªßa s·∫£n ph·∫©m C theo nƒÉm:")
print(df[["Year", "T·ª∑ tr·ªçng doanh thu C (%)", "T·ª∑ tr·ªçng gross profit C (%)", "T·ª∑ tr·ªçng l·ª£i nhu·∫≠n sau chi ph√≠ C (%)"]]
      .round(2).to_string(index=False))

# Tr·ª±c quan
plt.figure(figsize=(10,6))
plt.plot(df["Year"], df["T·ª∑ tr·ªçng doanh thu C (%)"], marker='o', label="Doanh thu")
plt.plot(df["Year"], df["T·ª∑ tr·ªçng gross profit C (%)"], marker='o', label="Gross Profit")
plt.plot(df["Year"], df["T·ª∑ tr·ªçng l·ª£i nhu·∫≠n sau chi ph√≠ C (%)"], marker='o', label="L·ª£i nhu·∫≠n sau chi ph√≠")

plt.title("T·ª∑ tr·ªçng ƒë√≥ng g√≥p c·ªßa s·∫£n ph·∫©m C trong c·∫•u tr√∫c t√†i ch√≠nh c·ªßa ITB")
plt.ylabel("T·ª∑ tr·ªçng (%)")
plt.xlabel("NƒÉm")
plt.ylim(0, 100)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ƒê·ªçc d·ªØ li·ªáu
product = pd.read_csv("product.csv")

# B·ªè d√≤ng t·ªïng v√† chuy·ªÉn Year sang int
product = product.dropna(subset=["Year"])
product["Year"] = product["Year"].astype(float).astype(int)

# X√≥a d·∫•u ph·∫©y v√† chuy·ªÉn c√°c c·ªôt sang float
for col in ['Sum of product_price', 'Gross Profit', 'Sum of ProfitOrder']:
    product[col] = product[col].astype(str).str.replace(",", "").astype(float)

# T√≠nh t·ªïng theo nƒÉm
total_by_year = (
    product.groupby("Year")[['Sum of product_price', 'Gross Profit', 'Sum of ProfitOrder']]
    .sum()
    .rename(columns={
        'Sum of product_price': 'Total_Revenue',
        'Gross Profit': 'Total_GrossProfit',
        'Sum of ProfitOrder': 'Total_NetProfit'
    })
)

# L·∫•y s·ªë li·ªáu s·∫£n ph·∫©m C
c_data = product[product['product_profile']=='product_C'].set_index('Year')

# T√≠nh t·ª∑ tr·ªçng (%)
df_ratio = pd.DataFrame({
    'Year': total_by_year.index,
    'Revenue_%': (c_data['Sum of product_price'] / total_by_year['Total_Revenue'] * 100).round(2),
    'GrossProfit_%': (c_data['Gross Profit'] / total_by_year['Total_GrossProfit'] * 100).round(2),
    'NetProfit_%': (c_data['Sum of ProfitOrder'] / total_by_year['Total_NetProfit'] * 100).round(2),
}).reset_index(drop=True)

# In b·∫£ng s·ªë li·ªáu r√µ r√†ng
print("T·ª∑ tr·ªçng ƒë√≥ng g√≥p c·ªßa s·∫£n ph·∫©m C theo nƒÉm (%):")
print(df_ratio.to_string(index=False))

# V·∫Ω bi·ªÉu ƒë·ªì c·ªôt nh√≥m
plt.figure(figsize=(10, 6))
bar_w = 0.2
x = range(len(df_ratio))
plt.bar([p - bar_w for p in x], df_ratio['Revenue_%'], width=bar_w, label='Doanh thu %')
plt.bar(x, df_ratio['GrossProfit_%'], width=bar_w, label='Gross Profit %')
plt.bar([p + bar_w for p in x], df_ratio['NetProfit_%'], width=bar_w, label='Net Profit %')
plt.xticks(x, df_ratio['Year'])
plt.ylabel('T·ª∑ tr·ªçng (%)')
plt.title('T·ª∑ tr·ªçng ƒë√≥ng g√≥p c·ªßa s·∫£n ph·∫©m C theo nƒÉm')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# B·ªè d√≤ng Total n·∫øu c√≤n s√≥t
df = sales_channel[sales_channel['sales_channel'] != 'Total'].copy()

# T·ªïng h·ª£p theo k√™nh
base = df.groupby('sales_channel').sum().reset_index()

# ================================================
# üîπ Scenario 1: Gi·∫£m 10% flexible_fee v√† commission cho t·∫•t c·∫£ k√™nh
# ================================================
sc1 = base.copy()
sc1['flexible_fee_new'] = sc1['Sum of flexible_fee'] * 0.9
sc1['commission_new'] = sc1['Sum of commission'] * 0.9
sc1['profit_new'] = (
    sc1['Sum of product_price']
    - sc1['Sum of product_cost']
    - sc1['Sum of FixedCost']
    - sc1['flexible_fee_new']
    - sc1['commission_new']
)
sc1['improve'] = sc1['profit_new'] - sc1['Sum of ProfitOrder']

# B·∫£ng k·∫øt qu·∫£ Scenario 1
res1 = sc1[['sales_channel', 'profit_new', 'improve']].rename(columns={
    'sales_channel': 'K√™nh',
    'profit_new': 'L·ª£i nhu·∫≠n m·ªõi',
    'improve': 'C·∫£i thi·ªán l·ª£i nhu·∫≠n'
})
display(res1)

# Bi·ªÉu ƒë·ªì Scenario 1
plt.figure(figsize=(8, 4))
sns.barplot(data=res1, x='K√™nh', y='C·∫£i thi·ªán l·ª£i nhu·∫≠n', palette='viridis')
plt.title('Scenario 1: Gi·∫£m 10% Flexible Fee & Commission')
plt.axhline(0, color='black', linestyle='--')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# ================================================
# üîπ Scenario 2: Gi·∫£m 20% commission cho k√™nh premium
# ================================================
sc2 = base.copy()
sc2['commission_new'] = sc2.apply(
    lambda row: row['Sum of commission'] * 0.8 if row['sales_channel'] == 'premium' else row['Sum of commission'],
    axis=1
)
sc2['profit_new'] = (
    sc2['Sum of product_price']
    - sc2['Sum of product_cost']
    - sc2['Sum of FixedCost']
    - sc2['Sum of flexible_fee']
    - sc2['commission_new']
)
sc2['improve'] = sc2['profit_new'] - sc2['Sum of ProfitOrder']

# B·∫£ng k·∫øt qu·∫£ Scenario 2
res2 = sc2[['sales_channel', 'profit_new', 'improve']].rename(columns={
    'sales_channel': 'K√™nh',
    'profit_new': 'L·ª£i nhu·∫≠n m·ªõi',
    'improve': 'C·∫£i thi·ªán l·ª£i nhu·∫≠n'
})
display(res2)

# Bi·ªÉu ƒë·ªì Scenario 2
plt.figure(figsize=(8, 4))
sns.barplot(data=res2, x='K√™nh', y='C·∫£i thi·ªán l·ª£i nhu·∫≠n', palette='magma')
plt.title('Scenario 2: Gi·∫£m 20% Commission (ch·ªâ Premium)')
plt.axhline(0, color='black', linestyle='--')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# ================================================
# üîπ Scenario 3: Chuy·ªÉn 10% doanh thu t·ª´ Ecommerce sang Website
# Chi ph√≠ chia t·ªâ l·ªá theo doanh thu m·ªõi
# ================================================
sc3 = base.copy()

# ƒêi·ªÅu ch·ªânh doanh thu
sc3['rev_new'] = sc3.apply(
    lambda row: row['Sum of product_price'] * (0.9 if row['sales_channel'] == 'ecommerce'
                                                else 1.1 if row['sales_channel'] == 'website'
                                                else 1.0),
    axis=1
)

# Chi ph√≠ ƒëi·ªÅu ch·ªânh theo t·ªâ l·ªá doanh thu m·ªõi
for col in ['Sum of product_cost', 'Sum of FixedCost', 'Sum of flexible_fee', 'Sum of commission']:
    sc3[f'{col}_new'] = sc3.apply(
        lambda row: row[col] * (row['rev_new'] / row['Sum of product_price']),
        axis=1
    )

# T√≠nh l·ª£i nhu·∫≠n m·ªõi
sc3['profit_new'] = (
    sc3['rev_new']
    - sc3['Sum of product_cost_new']
    - sc3['Sum of FixedCost_new']
    - sc3['Sum of flexible_fee_new']
    - sc3['Sum of commission_new']
)
sc3['improve'] = sc3['profit_new'] - sc3['Sum of ProfitOrder']

# B·∫£ng k·∫øt qu·∫£ Scenario 3
res3 = sc3[['sales_channel', 'profit_new', 'improve']].rename(columns={
    'sales_channel': 'K√™nh',
    'profit_new': 'L·ª£i nhu·∫≠n m·ªõi',
    'improve': 'C·∫£i thi·ªán l·ª£i nhu·∫≠n'
})
display(res3)

# Bi·ªÉu ƒë·ªì Scenario 3
plt.figure(figsize=(8, 4))
sns.barplot(data=res3, x='K√™nh', y='C·∫£i thi·ªán l·ª£i nhu·∫≠n', palette='cubehelix')
plt.title('Scenario 3: Chuy·ªÉn 10% Doanh thu t·ª´ Ecommerce ‚Üí Website')
plt.axhline(0, color='black', linestyle='--')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# üíæ Gi·∫£ s·ª≠ b·∫°n ƒë√£ c√≥ sales_channel.csv ƒë√£ ƒë·ªçc s·∫µn nh∆∞ sau:
# sales_channel = pd.read_csv("sales_channel.csv")
# B·ªè d√≤ng t·ªïng v√† chuy·ªÉn v·ªÅ float
sales_channel = sales_channel[sales_channel['sales_channel'] != 'Total'].copy()
for col in ['Sum of product_price','Sum of product_cost','Sum of FixedCost',
            'Sum of flexible_fee','Sum of commission','Sum of ProfitOrder']:
    sales_channel[col] = sales_channel[col].astype(str).str.replace(",", "").astype(float)

# T√≠nh t·ªïng theo k√™nh
base = sales_channel.groupby('sales_channel').sum().reset_index()

# ========================================
# üîπ Scenario: Gi·∫£m 10% flexible_fee & commission
# ========================================
sc1 = base.copy()
sc1['flexible_fee_new'] = sc1['Sum of flexible_fee'] * 0.9
sc1['commission_new'] = sc1['Sum of commission'] * 0.9
sc1['profit_new'] = (
    sc1['Sum of product_price']
    - sc1['Sum of product_cost']
    - sc1['Sum of FixedCost']
    - sc1['flexible_fee_new']
    - sc1['commission_new']
)
sc1['improve'] = sc1['profit_new'] - sc1['Sum of ProfitOrder']

# ===============================
# üìä So s√°nh t·ªïng l·ª£i nhu·∫≠n to√†n h·ªá th·ªëng
# ===============================
total_before = base['Sum of ProfitOrder'].sum()
total_after = sc1['profit_new'].sum()
total_improve = total_after - total_before

summary_df = pd.DataFrame({
    'Tr·∫°ng th√°i': ['Tr∆∞·ªõc m√¥ ph·ªèng', 'Sau m√¥ ph·ªèng (Gi·∫£m 10% FF & Comm)'],
    'T·ªïng l·ª£i nhu·∫≠n': [total_before, total_after]
})
display(summary_df)

# ===============================
# üìã Chi ti·∫øt t·ª´ng k√™nh
# ===============================
detailed_df = sc1[['sales_channel', 'Sum of ProfitOrder', 'profit_new', 'improve']].rename(columns={
    'sales_channel': 'K√™nh',
    'Sum of ProfitOrder': 'L·ª£i nhu·∫≠n ban ƒë·∫ßu',
    'profit_new': 'L·ª£i nhu·∫≠n m·ªõi',
    'improve': 'C·∫£i thi·ªán l·ª£i nhu·∫≠n'
})
display(detailed_df)

# ===============================
# üìà V·∫Ω bi·ªÉu ƒë·ªì t·ªïng l·ª£i nhu·∫≠n
# ===============================
plt.figure(figsize=(6, 4))
sns.barplot(data=summary_df, x='Tr·∫°ng th√°i', y='T·ªïng l·ª£i nhu·∫≠n', palette='Set2')
plt.title('T·ªïng l·ª£i nhu·∫≠n to√†n h·ªá th·ªëng tr∆∞·ªõc v√† sau m√¥ ph·ªèng')
plt.axhline(0, color='black', linestyle='--')
plt.tight_layout()
plt.show()

# ===============================
# üìâ V·∫Ω bi·ªÉu ƒë·ªì l·ª£i nhu·∫≠n theo t·ª´ng k√™nh
# ===============================
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.barplot(data=detailed_df, x='K√™nh', y='L·ª£i nhu·∫≠n ban ƒë·∫ßu', ax=ax[0], color='skyblue')
ax[0].set_title('L·ª£i nhu·∫≠n ban ƒë·∫ßu theo t·ª´ng k√™nh')
ax[0].tick_params(axis='x', rotation=30)

sns.barplot(data=detailed_df, x='K√™nh', y='L·ª£i nhu·∫≠n m·ªõi', ax=ax[1], color='lightgreen')
ax[1].set_title('L·ª£i nhu·∫≠n sau khi gi·∫£m 10% ph√≠ v·∫≠n chuy·ªÉn & hoa h·ªìng')
ax[1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```
